```python
from xbbg import blp
import pandas as pd
data = pd.read_excel("C:\dataset\ISO10383_MIC.xls")
MIC = data["Column1"]
MIC = list(MIC)
```

```python
blp.bdp(tickers=MIC, flds=['Security_Name', 'GICS_Sector_Name'])
```

```python
risk = blp.bdh(tickers=MIC, flds=['EQY_BETA','EQY_ALPHA','SPREAD_CNVX_ASK','HEDGE_GAP','INDEX_Z_SPREAD',
                                        'BS_TOT_CAP_TO_RISK_BASE_CAP','BS_RISK_WEIGHTED_ASSETS','BS_TOTAL_RISKS',
                                        'ARDR_TOTAL_VALUE_AT_RISK','IS_TOTAL_VALUE_AT_RISK','VALUE_AT_RISK_TO_TOTAL_EQUITY',
                                        'VOLATILITY_90D','WACC_COST_EQUITY','RISK_PREMIUM','ALTMAN_Z_SCORE',
                                        'EQY_SHARPE_RATIO_3YR','LQA_PRICE_VOLATILITY','ATT_CORR_SEN','MLI_EFF_RISK',
                                        'STDDEV_ERROR','EQY_BETA_6M','EQY_ALPHA_6M','CORR_COEF','YAS_RISK',
                                        'WACC_COST_DEBT'], start_date='2010')
print(risk)
```
